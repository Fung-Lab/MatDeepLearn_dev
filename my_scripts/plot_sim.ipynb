```python
import numpy as np
```

```python
def get_records(list_i):
    content = []
    for i in list_i:
        content.append(np.load(f'../train_outs/sim_records/energy_rdf_min_dists_{i}000000.npz'))
    return content
```

```python
records = get_records(range(1, 8, 1))
records[0].keys()
```

```python
energy = np.concatenate([records[i]['energy'] for i in range(len(records))])
rdf = np.concatenate([records[i]['rdf_changes'] for i in range(len(records))])
min_dists = np.concatenate([records[i]['min_dists'] for i in range(len(records))])
```

```python
timesteps = np.arange(0, energy.shape[0], 1)
```

```python
splited = np.split(energy, 8)
top_energy_indices = np.array([np.argmax(arr) + i * splited[0].shape[0] for i, arr in enumerate(splited)])
top_energy_indices
```

```python
import matplotlib.pyplot as plt
import numpy as np

top_energy_indices = np.array([3945, 5066, 5255])
top_energy_timesteps = timesteps[top_energy_indices]

fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(8, 10))

ax1.plot(timesteps, energy)
ax1.set_xlabel('Timesteps')
ax1.set_ylabel('Energy')
ax1.set_title('Energy vs. Timesteps')
for i, timestep in enumerate(top_energy_timesteps, start=1):
    ax1.axvline(timestep, color=f'C{i}', linestyle='--', label=f'Top {i} Energy at Timestep {timestep}')


ax2.plot(timesteps, rdf)
ax2.set_xlabel('Timesteps')
ax2.set_ylabel('RDF')
ax2.set_title('RDF vs. Timesteps')
for i, timestep in enumerate(top_energy_timesteps, start=1):
    ax2.axvline(timestep, color=f'C{i}', linestyle='--', label=f'Top {i} Energy at Timestep {timestep}')


ax3.plot(timesteps, min_dists)
ax3.set_xlabel('Timesteps')
ax3.set_ylabel('Minimum Distances')
ax3.set_title('Minimum Distances vs. Timesteps')
for i, timestep in enumerate(top_energy_timesteps, start=1):
    ax3.axvline(timestep, color=f'C{i}', linestyle='--', label=f'Top {i} Energy at Timestep {timestep}')


plt.tight_layout()
plt.show()
```
